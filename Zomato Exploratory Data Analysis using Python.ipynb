```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

ratings

##### Observation 1:Whenever the rating is from 4.5 to 4.9,it indicates that it is excellent
Observation 2:Whenever the rating is from 4.0 to 4.5,it indicates that it is very good.
Observation 3:Whenever the rating is from 3.5 to 3.9,it indicates that it is good.
Observation 4:Whenever the rating is from 3.0 to 3.4,it indicates that it is average.
Observation 5:Whenever the rating is from 2.5 to 2.9,it indicates that it is poor.

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

### Observation:

    2.Maximum number of rating are between 2.5 to 3.4
